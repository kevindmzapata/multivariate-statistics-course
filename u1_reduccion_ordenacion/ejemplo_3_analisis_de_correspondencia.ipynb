```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import prince
```

```python
df = pd.read_csv("databases/correspondencia.csv", sep=";", index_col=0)
df.info()
```

```python
columns = ["Carrera", "Genero", "Estrato", "Origen", "Edad"]
df.columns = columns
df.head()
```

```python
# Quitar edad
df = df.drop("Edad", axis=1)
df.head()
```

```python
mca = prince.MCA(
    n_components=2,
    n_iter=3,
    copy=True,
    check_input=True,
    engine='sklearn',
    random_state=42
)

mca = mca.fit(df)
```

```python
one_hot = pd.get_dummies(df)

mca_no_one_hot = prince.MCA(one_hot=False)
mca_no_one_hot = mca_no_one_hot.fit(one_hot)
```
